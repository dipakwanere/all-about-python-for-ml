#### Visualization

- In this notebook, I'll be showing how wee can use the different visualizations for creating graphs, charts and some amazing features for dashboards.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(50,60)
```

```python
#plotting using matplotlib
plt.scatter(x,y,c="g")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Graph in 2D")
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*--',linestyle='dashed',linewidth=2,markersize=12)
```

```python
# plotting subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```
