#### Visualization

- In this notebook, I'll be showing how wee can use the different visualizations for creating graphs, charts and some amazing features for dashboards.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(50,60)
```

```python
#plotting using matplotlib
plt.scatter(x,y,c="g")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Graph in 2D")
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*--',linestyle='dashed',linewidth=2,markersize=12)
```

```python
# plotting subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
# plots in seaborn, 
# bivariate - distplot, joinplot,pairplot
```

```python
df.describe()
```

```python
# Finding the correlation - works with numerical or flot features not with categorical features
corr_matrix= df.corr(numeric_only=True)
corr_matrix
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
# Univariate analysis with joinplot (uses 2 features)
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
# When you have more than 2 featues - use pairplaot (its a scatter plot with other features)
sns.pairplot(data=df)
```

```python
sns.pairplot(data=df,hue='sex')
```

```python
# Dist plot - 
```

```python
sns.displot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```
