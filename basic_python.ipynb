```python
# TODO: Edit the function to return the correct grade for different scores
def get_grade(score):
    if score >=90 and score<=100:
        return "A"
    elif score >=80 and score<90:
        return "B"
    elif score >=70 and score<80:
        return "C"
    elif score >=60 and score<70:
        return "D"
    else:
        return "F"
```

```python
# Generator - a generator is a special kind of iterator that
#              generates items on the fly instead of storing them in memory.

# - Saves memory for large datasets
# - Defined like a function but uses yeild instead of return
```

```python
# cerate Generator
def generate_square(n):
    for i in range(1,n+1):
        yield i*i

# create generator
square_generator =  generate_square(5)
 

 # get values one by one with -- next()
print(next(square_generator)) 
```

```python
# memory efficient example:
# large range without storing all numbers
def generate_numbers(limit):
    for i in range(limit):
        yield i

for num in generate_numbers(1000000):
    if num < 5:
        print(num)
    else:
        break
```

```python
# Decorators - A decorator is a function that wraps another function to add extra functionality without changing the original function.

# Analogy:
# - think of gift wrapping a present
# - original gift = original function
# - wrapping paper = decorator (adds style/behaviour)
# - Gift inside stays the same
```

```python
# Basic decorator function

def announce_decorator(func):
    def wrapper(*args):
        print("this is coming from a decorator function")
        func(*args)
        print("function as finished running..")
    return wrapper
```

```python
# original function
def say_hello():
    print("Hi, I am original function...!")
```

```python
# decorate manually
decorated_function = announce_decorator(say_hello)
decorated_function()
```

```python
# using @ syntax (pythonic way)

@announce_decorator
def greet():
    print("Greetings, Dipak!")

greet()
```

```python
# Decorator with arguments 

def repeat_decorator(times):
    def decorator(func):
        def wrapper(*args,**kwargs):
            for _ in range(times):
                func(*args,**kwargs)
        return wrapper
    return decorator

@repeat_decorator(times=5)
def say_hi(name):
    print(f"Hi, {name}")
```

```python
say_hi("pintooo")
```

```python
# Decorator with arguments 
def repeat_decorator(func):
    def wrapper(*args, **kwargs):
        func(*args, **kwargs)
    return wrapper

@repeat_decorator
def cal_sum(x, y):
    print(x + y)
```

```python
cal_sum(10, 20)
```
